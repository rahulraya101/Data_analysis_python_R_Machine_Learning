```python
import csv

date = []
all_transaction = []
sum_balance = []
with open('/home/rahulraya/Desktop/Python_R_Data_Analysis/validation_input.csv', encoding="utf8") as csv_file:
   csv_reader = csv.reader(csv_file, delimiter=';')
   line_count = 0
   balance = 0
   for row in csv_reader:
       date.append(row[0])
       transaction = int(row[1])
       all_transaction.append(transaction)
       #print(f'date is {date} and transaction is {transaction}')        
       balance = balance + transaction
       sum_balance.append(balance)
   #print(f'balance is {balance}') # plot transaction vs each balance,
print(f'transaction is {all_transaction}')
print(f'balance is {sum_balance}')
```

```python
fig, ax = plt.subplots(figsize=(16, 4))
ax.plot(
    all_transaction,
    sum_balance,
    linestyle='none',
    marker='x')
#fig.autofmt_xdate()
ax.set_xlabel('transaction')
ax.set_ylabel('balance')
plt.show()
```
