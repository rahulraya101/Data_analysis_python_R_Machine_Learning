```r
#Q_1 Prepare a function converting degrees of Fahrenheit to Centigrade
fahrenheit <- c()
centirgrade <- (fahrenheit - 32) * (5 / 9)
```

```r
#Q_2 Debug the function using the values provided in above figure — how much is in Celsius –40, 98.6, 212 °F?
fahrenheit <- c(-40 , 98.6, 212)
centirgrade <- c(fahrenheit - 32) * (5 / 9)
conversion <- data.frame( Fahrenheit=fahrenheit, Centirgrade =centirgrade)
print(conversion)
```

```r
#Q_3 Design a conversion table Fahrenheit to Celsius for 0, 10, …220 °F and plot the data into a binary diagram
fahrenheit <- 0:220
centirgrade <- (fahrenheit - 32) * (5 / 9)
conversion <- data.frame(Fahrenheit=fahrenheit, Centirgrade =centirgrade )
print(conversion)
```

```r
plot(centirgrade,  fahrenheit, main = "plot(Centirgrade,  Fahrenheit)")
```

```r
fahrenheit <- c(0:220)

Convert2Celsius <- function(fahrenheit)
{
centirgrade <- (fahrenheit - 32) * 5/9
return (centirgrade)
}

Convert2Celsius(fahrenheit)
```

```r
library(data.table)

convert.table <- data.table(Celsius = centirgrade, Fahrenheit = fahrenheit)
print(convert.table)
```

```r
# define temperature as temp
f.temp <-c(0:220)

# Returns Celsius conversion of passed f.temp (in Fahrenheit)
convert.to.cel <- function(f.temp)
{
c.temp <- (f.temp - 32) * 5/9
return(c.temp)
}

convert.to.cel(f.temp)
```

```r
#library(data.table)
convert.table <- data.table(celsius = c.temp, fahrenheit = f.temp)
print(convert.table)
```

```r
# plot the data into a binary diagram
plot(convert.table,
     main = "conversion table",
     type = "l",
     col = "green",
     pch = 16,
     xlab = "centigrade",
     ylab = "fahrenheit")
```
